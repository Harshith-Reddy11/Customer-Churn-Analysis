```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('/content/Customer Churn.csv')
df.head()
```

```python
df.info()
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ','0')
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
def conv_value(value):
  if value == 1:
    return 'yes'
  else:
    return 'no'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv_value)
df.head()
```

```python
ax = sns.countplot(x ='Churn' , data =df)
ax.bar_label(ax.containers[0])
plt.title('Churn Distribution')
plt.show()
```

```python
plt.figure(figsize = (10,5))
gb = df.groupby('Churn').agg({'Churn' : 'count'})
gb
plt.pie(gb['Churn'] , labels = gb.index , autopct = "%1.2f%%")
plt.title('Percentage of churned cutomers')
plt.show()
```

```python
 sns.countplot(x=df['gender'], hue=df['Churn'],palette='inferno')
 plt.title('Gender Distribution')
 plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['SeniorCitizen'],hue=df['Churn'],palette='inferno')
plt.title('Senior Citizen Distribution')
plt.show()
```

```python
# Calculate the percentage distribution for each group
churn_percentage = df.groupby(['SeniorCitizen', 'Churn']).size().unstack().fillna(0)
churn_percentage = churn_percentage.div(churn_percentage.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart using pandas built-in plotting
ax = churn_percentage.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='viridis')

# Adding labels and title
ax.set_title('Senior Citizen Churn Distribution as Percentage', fontsize=14, color='white')
ax.set_xlabel('Senior Citizen', fontsize=12, color='white')
ax.set_ylabel('Percentage (%)', fontsize=12, color='white')

# Adding percentage labels on top of the bars
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x, y = p.get_xy()  # Get the x and y coordinates of the bar
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center', color='white')

# Set the background color of the plot to dark
ax.set_facecolor('#212121')  # Dark background color
ax.figure.patch.set_facecolor('#212121')  # Dark background color for figure

# Adjust gridlines to be more visible on dark backgrounds
# ax.grid(True, color='white', linestyle='--', alpha=0.3)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(x=df['tenure'],bins=50,hue=df['Churn'],palette='inferno')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['Contract'],hue=df['Churn'],palette='inferno')
plt.title('Count of cutomers by contract')
plt.show()
```

```python
df.columns.values
```

```python
# Define columns to plot
columns_to_plot = ['PhoneService', 'MultipleLines', 'InternetService',
                   'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                   'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 15))  # Increased figure size for better readability
axes = axes.flatten()  # Flatten for easy indexing

# Loop through the columns and plot
for i, col in enumerate(columns_to_plot):
    sns.countplot(x=df[col], hue=df['Churn'], palette='inferno', ax=axes[i])
    axes[i].set_title(col, fontsize=12, color='white')
    axes[i].tick_params(axis='x', rotation=360, colors='white')
    axes[i].tick_params(axis='y', colors='white')
    axes[i].set_facecolor('#212121')
    axes[i].legend(title='Churn', loc='upper right', facecolor='#212121', edgecolor='white', labelcolor='white')


# Remove empty subplot if fewer than total (not needed here as we have 9 columns and 9 subplots)
# for j in range(i + 1, len(axes)):
#     fig.delaxes(axes[j])

# Figure styling
fig.patch.set_facecolor('#212121')  # Background
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = sns.countplot(x=df['PaymentMethod'],hue=df['Churn'],palette='inferno')
for container in ax.containers:
    ax.bar_label(container)
plt.title('Count of cutomers by PaymentMethod')
plt.xticks(rotation = 45)
plt.show()
```
